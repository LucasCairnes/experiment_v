```python
import numpy as np

grad_raw = 657
D_cm = 1.637
a_cm = 2.387
b_cm = 1.764

Delta_grad_raw = 4
Delta_D_cm = 0.005
Delta_a_cm = 0.008
Delta_b_cm = 0.007

g_cgs = 981.0

conversion_factor = 10.0 / 1000.0
grad_cgs = grad_raw * conversion_factor
Delta_grad_cgs = Delta_grad_raw * conversion_factor

numerator_mu = grad_cgs * g_cgs * D_cm * (a_cm**2 - b_cm**2)
denominator_mu = 8 * (np.pi**2) * (a_cm**2) * (b_cm**2)
mu_poise = numerator_mu / denominator_mu

term1_grad = (Delta_grad_cgs / grad_cgs)**2

term2_D = (Delta_D_cm / D_cm)**2

term3_a_num = 2 * (b_cm**2) * Delta_a_cm
term3_a_den = a_cm * (a_cm**2 - b_cm**2)
term3_a = (term3_a_num / term3_a_den)**2

term4_b_num = 2 * (a_cm**2) * Delta_b_cm
term4_b_den = b_cm * (a_cm**2 - b_cm**2)
term4_b = (term4_b_num / term4_b_den)**2

fractional_error_sum = term1_grad + term2_D + term3_a + term4_b

Delta_mu_poise = mu_poise * np.sqrt(fractional_error_sum)

poise_to_pascal_seconds = 0.1
mu_pa_s = mu_poise * poise_to_pascal_seconds
Delta_mu_pa_s = Delta_mu_poise * poise_to_pascal_seconds

print("\n--- Final Results (in SI units) ---")
print(f"Viscosity (μ):     {mu_pa_s:.4f} Pa·s")
print(f"Absolute Error (Δμ): {Delta_mu_pa_s:.4f} Pa·s")
print("---------------------------------------------")
print(f"Final Value: μ = {mu_pa_s:.4f} ± {Delta_mu_pa_s:.4f} Pa·s")
```
