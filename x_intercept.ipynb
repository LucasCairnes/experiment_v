```python
import math

def get_x_intercept(m, m_err, c, c_err):
    x = -c / m
    x_err = (1 / abs(m)) * math.sqrt(c_err**2 + (x * m_err)**2)
    return x, x_err

# Inputs
slope = 150
slope_err = 10
y_int = 1100
y_int_err = 300

val, err = get_x_intercept(slope, slope_err, y_int, y_int_err)

print(f"{val:.4f} ± {err:.4f}")
```
